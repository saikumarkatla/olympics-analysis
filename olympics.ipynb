```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top10_combined, x="Combined Total", y="Team", palette="viridis")
plt.title("Top 10 Teams by Combined Total Medals")
plt.xlabel("Total Medals")
plt.ylabel("Team")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top10_summer, x="Summer Total", y="Team", palette="mako")
plt.title("Top 10 Teams by Summer Total Medals")
plt.xlabel("Total Summer Medals")
plt.ylabel("Team")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top10_winter, x="Winter Total", y="Team", palette="coolwarm")
plt.title("Top 10 Teams by Winter Total Medals")
plt.xlabel("Total Winter Medals")
plt.ylabel("Team")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df[["Summer Gold", "Winter Gold", "Combined Gold"]].sum().plot(kind="bar", color=["gold", "skyblue", "orange"])
plt.title("Gold Medal Distribution (Overall)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df[["Summer Silver", "Winter Silver", "Combined Silver"]].sum().plot(kind="bar", color=["grey", "lightblue", "blue"])
plt.title("Silver Medal Distribution (Overall)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df[["Summer Bronze", "Winter Bronze", "Combined Bronze"]].sum().plot(kind="bar", color=["brown", "lightcoral", "darkred"])
plt.title("Bronze Medal Distribution (Overall)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.drop(columns="Team").corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Medal Counts")
plt.show()
```

```python
df["Summer_vs_Winter"] = df["Summer Total"] - df["Winter Total"]
```

```python
plt.figure(figsize=(10,6))
top_diff = df.nlargest(10, "Summer_vs_Winter")
sns.barplot(data=top_diff, x="Summer_vs_Winter", y="Team", palette="plasma")
plt.title("Top 10 Teams with Higher Summer Performance")
plt.xlabel("Summer - Winter Medal Difference")
plt.ylabel("Team")
plt.show()
```

```python
bottom_diff = df.nsmallest(10, "Summer_vs_Winter")
plt.figure(figsize=(10,6))
sns.barplot(data=bottom_diff, x="Summer_vs_Winter", y="Team", palette="cubehelix")
plt.title("Top 10 Teams with Higher Winter Performance")
plt.xlabel("Summer - Winter Medal Difference")
plt.ylabel("Team")
plt.show()
```
